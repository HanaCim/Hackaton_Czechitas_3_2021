```python
import glob
import pandas as pd
```

```python
files = glob.glob("data/*.csv")
```

```python
files
```

```python
names = []
for i, file in enumerate(files):
    name = "file" + str(i + 1)
    name = pd.read_csv(file, header=9, names=('Date', 'Time', 'ServerWatchdog', 'BeamIsOn', 'ExternalState', 'IsBlanked', 'ServiceModeIsOn'), parse_dates=[['Date', 'Time']])
    names.append(name)
    
```

```python
names2 = []
for name in names:
    name = name.fillna(method='ffill')
    name = name.dropna()
    names2.append(name)

    print(name.info())
    print(name.head())
    
```

```python
names2[0]
```

```python
names3 = []
for name in names2:
    name['Date_Time'] = pd.to_datetime(name['Date_Time'])
    names3.append(name)
```

```python
for name in names3:
    print(name.info())
```

```python
names4 = []
for i, name in enumerate(names3):
    name['MicroscopeNro'] = i + 1
    names4.append(name)
```

```python
names4
```

```python
final = pd.concat([name for name in names4])
```

```python
final
```

```python
final = final.reset_index()
```

```python
final
```

```python
# OnOff column

final.loc[(final['ServiceModeIsOn'] == 'On (1)'), 'OnOff'] = 'Service'  
final.loc[(final['ServerWatchdog'] == 1) & (final['BeamIsOn'] == 'On (1)') & (final['ExternalState'] == 'Maintain (4)') & (final['IsBlanked'] ==  0) & (final['ServiceModeIsOn'] == 'Off (0)'), 'OnOff'] = 'Used' 
final.loc[(final['ServerWatchdog'] == 1) & (final['BeamIsOn'] == 'On (1)') & (final['ExternalState'] == 'Maintain (4)') & (final['IsBlanked'] ==  1) & (final['ServiceModeIsOn'] == 'Off (0)'), 'OnOff'] = 'On'

                            
                            
```

```python
final.loc[(final['OnOff'] != 'Service') & (final['OnOff'] != 'Used') & (final['OnOff'] != 'On'), 'OnOff'] = 'Off'
```

```python
final = final.drop(['IsUsed'], axis=1)
```

```python
final.head(50)
```

```python
final.to_csv('microscopes.csv', index=False)
```

```python
# Off=1
# On=2
# Used=3
# Service=4
final.loc[(final['OnOff'] == 'Off'), 'OnOff'] = 1
final.loc[(final['OnOff'] == 'On'), 'OnOff'] = 2
final.loc[(final['OnOff'] == 'Used'), 'OnOff'] = 3
final.loc[(final['OnOff'] == 'Service'), 'OnOff'] = 4
```

```python
final = final.loc[final['OnOff'].diff() != 0]
```

```python
final
```

```python
final['Difference'] = final['Date_Time'].diff()
```
